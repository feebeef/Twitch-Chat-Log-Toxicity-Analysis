```python
labels = ["Direct Harassment", "Hate Speech","Sexual Harassment","Trolling", "Others", "Toxic"]

batch_files = ['usanews.csv', 'foxnews0.csv', 'foxnews1.csv', 'foxnews2.csv', 'foxnews3.csv', 
                'nogla0.csv','pew0.csv', 'rae0_0.csv', 
                 'rae0_1.csv', 'terror0_0.csv', 'terror0_1.csv', 'disrespect0_0.csv', 'disrespect0_1.csv']


vid_files = ["foxnews_FULL.csv", "rae_FULL.csv",  "terror_FULL.csv",
             "usanews.csv" , "pew0.csv", "nogla0.csv", "disrespect_FULL"]


genre_files = ["gaming_channels_GENRE.csv", "news_channel_GENRE.csv" ]

```

```python
src_folder = "Annotations/CSVs/"
dest_folder = "Annotations/CSVs/"
files = batch_files
for file in batch_files:
    df = pd.read_csv(src_folder + file)
    df = df.loc[:, ~df.columns.str.contains('^Unnamed:')]
    df[labels] = df[labels].astype("float").replace(r'^\s*$', np.nan, regex=True).fillna(0).replace(np.nan,0)
    df["None"] = (df[labels].max(axis=1) == 0).astype(float)
    df.to_csv(dest_folder + file)
```

```python
import pandas as pd
import numpy as np
import nltk

src_folder = "Annotations/CSVs/"
dest_folder = "Annotations/CSVs/"

def combine_batch(files, name):
    dfs = []
    new_columns = [] 
    for file in files:
        df = pd.read_csv(src_folder + file, index_col=0)
        dfs.append(df)
    combined_df = pd.DataFrame(columns=dfs[0])
    combined_df = pd.concat(dfs, axis=0, ignore_index=True)
    combined_df[combined_df.columns.difference(labels)].replace(np.nan, '', regex=True)
    combined_df.to_csv( dest_folder + name )
    return combined_df
```

```python
foxnews = ['foxnews0.csv', 'foxnews1.csv', 'foxnews2.csv', 'foxnews3.csv']
rae = ['rae0_0.csv', 'rae0_1.csv']
terror = ['terror0_0.csv', 'terror0_1.csv']
dr = ['disrespect0_0.csv', 'disrespect0_1.csv']

display("Combined Channels")
display(combine_batch(foxnews, "foxnews_FULL.csv").info())
display(combine_batch(rae, "rae_FULL.csv").info())
display(combine_batch(terror, "terror_FULL.csv").info())
display(combine_batch(dr, "drdisrespect_FULL.csv").info())
```

```python
gaming_channels = ["rae_FULL.csv",  "terror_FULL.csv", "nogla0.csv","pew0.csv", "drdisrespect_FULL.csv"]
news_channel = ["foxnews_FULL.csv", "usanews.csv"]

display("Combined Genre")
display(combine_batch(gaming_channels, "gaming_channels_GENRE.csv").info())
display(combine_batch(news_channel, "news_channel_GENRE.csv").info())
```
