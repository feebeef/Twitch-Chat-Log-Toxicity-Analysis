### Import Libraries 

```python
import pandas as pd
import base64
import random
from operator import itemgetter
import json
```

### Custom Functions

```python
anonDB = {}
anonFormat = "username@"
def anonymize(text):
    chars=[]
    if text in anonDB:
        return anonDB.get(text)
    else:
        for i in range(10):
            chars.append(random.choice(text))
        uid = "".join(chars)
        anonDB[text] = anonFormat+uid
    return anonFormat + uid
```

### Clean CSV

```python
fileName = "foxnews2" # <----- CHANGE VALUE PER FILE
df = pd.read_csv(fileName + ".csv")
df.head(10)
```

```python
df.drop(df.columns.difference(['author','message', 'time_text']), 1, inplace=True)
df.head(50)
```

```python
df['author'] = df['author'].apply(anonymize)
df.head(50)
```

```python
df["Direct Harassment"] = ""
df["Hate Speech"] = ""
df["Sexual Harassment"] = ""
df["Trolling"] = ""
df["Others"] = ""
df["Toxic"] = ""

df.head()
```

```python
df.to_csv(fileName + "_cleanedAnon.csv", index=True)
db = fileName + "anonymizedKeys.json"
with open(db, 'w') as fp:
    json.dump(anonDB, fp, indent=4)
```
