### Import Libraries 

```python
import pandas as pd
import base64
import random
from operator import itemgetter
import json
import re
```

### Custom Functions

```python
anonDB = {}
anonFormat = "username@"
def anonymize(text):
    chars=[]
    if text in anonDB:
        return anonDB.get(text)
    else:
        for i in range(10):
            chars.append(random.choice(text))
        uid = "".join(chars)
        anonDB[text] = anonFormat+uid
    anon = anonFormat+uid
    anone = re.sub('\s', '', anon)
    return anon

def anonymizeTags(df):
    names = list(anonDB.keys())
    for name in names:
        tag = '@' + name
        #print(tag)
        anon = '['+anonDB[name]+']'
        df = df.replace(to_replace=tag, value=anon, regex=True)
        
    df = df.replace(to_replace='(?<!username)@[A-Za-z]*', value='', regex=True)
    return df
        
    
```

### Clean CSV

```python
fileName = "rae" # <----- CHANGE VALUE PER FILE
df = pd.read_csv(fileName + ".csv")
df.head(10)
```

```python
df.drop(df.columns.difference(['author','message', 'time_text']), 1, inplace=True)
df.head(50)
```

```python
df['author'] = df['author'].apply(anonymize)
df.head(50)
```

```python
df = anonymizeTags(df)
df.head(20)
```

```python
df["Direct Harassment"] = ""
df["Hate Speech"] = ""
df["Sexual Harassment"] = ""
df["Trolling"] = ""
df["Others"] = ""
df["Toxic"] = ""

df.head()
```

```python
df.to_csv(fileName + "_cleanedAnon.csv", index=True)
db = fileName + "anonymizedKeys.json"
with open(db, 'w') as fp:
    json.dump(anonDB, fp, indent=4)
```
