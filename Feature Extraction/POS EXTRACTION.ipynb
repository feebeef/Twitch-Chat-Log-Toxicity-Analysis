# POS Tagging

```python
import spacy
from spacymoji import Emoji
import contractions
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
nlp = spacy.load("en_core_web_sm")
emoji = Emoji(nlp)
nlp.add_pipe('emoji', first=True)   

def expand_contractions(message):
    words = []
    for word in list(message.split(" ")):
        word = contractions.fix(word)
        words.append(word)
    return " ".join(words)

def pos_tagging(message):
    doc = nlp(message)
    result = []
    for token in doc:
        if token._.is_emoji:
            result.append((token.lemma_, "EMOJI"))
        elif token.lemma_ in nlp.vocab:
            result.append((token.lemma_, token.pos_))
        else:
            result.append(("UKNOWN", token.pos_))
    return result

# test["NEW_POS"] = test['cleaned_message'].apply(lambda message: pos_tagging(message))
# display(test["NEW_POS"][1])
```

```python
src_folder = "../Annotations/CSVs/"
dest_folder = "../Annotations/CSVs/"

for file in genre_files: 
    df = pd.read_csv(src_folder + file, index_col = 0 )
    df["cleaned_message"].replace(np.nan, '', regex=True, inplace = True)
    df["POS"] = df["cleaned_message"].apply(lambda message: expand_contractions(message))
    df["POS"] = df["POS"].apply(lambda message: pos_tagging(message))
    df.to_csv(dest_folder + file)
```

# POS SEQUENCE

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop = stopwords.words('english')

def pos_sequence(tokens):
    tags = []
    for word, tag in tokens:
        if not word in stop:
            if not (tag == "PUNCT" or tag == "AUX" or tag == "CONJ" or tag == "CCONJ" or tag == "DET" or tag == "SCONJ" 
                    or tag == "SYM" or tag == "ADP" or tag == "adp") :
                tags.append(tag)
    return " ".join(tags)
```

```python
import pandas as pd
src_folder = "../Annotations/CSVs/"
dest_folder = "../Annotations/CSVs/"

for file in files: 
    df = pd.read_csv(src_folder + file, index_col = 0 )
    df["POS"].replace(np.nan, '', regex=True, inplace = True)
    df["POS-SEQ"] = df["POS"].apply(lambda x: pos_sequence( eval(x)))
    df.to_csv(dest_folder + file)
```
