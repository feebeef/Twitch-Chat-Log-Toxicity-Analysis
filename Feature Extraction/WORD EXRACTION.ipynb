# PRE PROCESSING

```python
import emoji
import string
import re
from nltk.stem import WordNetLemmatizer as wordnet_lemmatizer
from nltk.corpus import wordnet
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

def lemmatize(word, tag):
    if tag == 'NOUN':
        return wordnet_lemmatizer.lemmatize(wordnet.NOUN, word)
    elif tag == 'VERB':
        return wordnet_lemmatizer.lemmatize(wordnet.VERB, word)
    elif tag == 'ADJ':
        return wordnet_lemmatizer.lemmatize(wordnet.ADJ, word)
    elif tag == 'ADV':
        return wordnet_lemmatizer.lemmatize(wordnet.ADV, word)
    else:
        return word
    
def extract_words(tokens):
    words = []
    for word, tag in tokens:
        if not tag == "EMOJI":
            words.append(lemmatize(word, tag))
    words = [w for w in words if not w in string.punctuation]
    words = [w for w in words if not w.lower() in stop_words]
    return " ".join(words)
```

# UPPERCASE WORD COUNT

```python
def count_upper_case(message):
    count = 0
    m = " ".join(re.findall("[a-zA-Z]+", message))
    words = list(m.split(" "))
    for word in words:
        if word.isupper():
            count+=1
    return count

print(count_upper_case("HELLO !!!! WHAT IS IT poop sddasdasd "))
```

```python
src_folder = "../Annotations/CSVs/"
dest_folder = "../Annotations/CSVs/"

for file in files:
    df = pd.read_csv(src_folder + file, index_col=0)
    df['uppercase_word_count'] = df['cleaned_message'].apply(lambda message: count_upper_case(str(message)))
    df.to_csv(dest_folder + file)
for file in files:
    df = pd.read_csv(dest_folder + file, index_col=0)
    display(df.head())
```

# PUNCTUATION COUNT

```python
from string import punctuation
from collections import Counter
def count_punctuations_symbol(message):
    count = 0
    for c in list(message):
        if c in punctuation:
            count += 1
    return count
print(count_punctuations_symbol("HELLO !!!! WHAT IS IT poop sddasdasd "))
```
