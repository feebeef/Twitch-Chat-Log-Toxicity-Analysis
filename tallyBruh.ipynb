```python
print("VIDEO FILES RATIO\n")
for file in vid_files:
    df = pd.read_csv(src_folder+file)
    df = df.drop(df.columns.difference(['Direct Harassment','Hate Speech', 'Sexual Harassment', 'Trolling', 'Others', 'Toxic']), 1, inplace=False)
    df['sum'] = df.sum(axis=1)
    val1=(df['sum'] > 0.0).sum()
    val2=df.shape[0]

    ratio = val1/val2
    percentage = "{:.0%}".format(ratio)
    print("Ratio in "+file+ ": " + percentage + "\tRatio exact Value: "+ str(ratio)+"\t"+str(val1)+"/"+str(val2))
```

```python
print("GENRE FILES RATIO\n")
for file in genre_files:
    df = pd.read_csv(src_folder+file)
    df = df.drop(df.columns.difference(['Direct Harassment','Hate Speech', 'Sexual Harassment', 'Trolling', 'Others', 'Toxic']), 1, inplace=False)
    df['sum'] = df.sum(axis=1)
    val1=(df['sum'] > 0.0).sum()
    val2=df.shape[0]

    ratio = val1/val2
    percentage = "{:.0%}".format(ratio)
    print("Ratio in "+file+ ": " + percentage + "\tRatio exact Value: "+ str(ratio)+"\t"+str(val1)+"/"+str(val2))
```
